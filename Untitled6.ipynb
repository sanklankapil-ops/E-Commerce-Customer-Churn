```python
# FINAL: Robust pie-chart plotting (no overlapping numbers, no syntax errors)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- CONFIG ---
cat_vars = ['CityTier', 'HourSpendOnApp', 'NumberOfDeviceRegistered', 'SatisfactionScore']
figsize = (24, 18)          # wide so legends fit
min_pct_inside = 3.0        # only show percentages inside pie when >= this value
explode_threshold_pct = 5.0 # explode slices smaller than this percent for visibility

# --- SAFETY CHECK ---
if not isinstance(df, pd.DataFrame):
    raise RuntimeError("df not found or not a pandas DataFrame. Load your dataframe into variable `df` before running this cell.")
missing = [c for c in cat_vars if c not in df.columns]
if missing:
    raise KeyError(f"Missing columns in df: {missing}")

# Optional: combine very small HourSpendOnApp categories into "Other" to reduce slices
def group_small(series, thresh_pct=0.01):
    counts = series.value_counts()
    total = counts.sum()
    small = counts[counts/total < thresh_pct]
    if len(small) == 0:
        return series
    return series.replace({k: 'Other' for k in small.index.tolist()})

# Apply grouping only to HourSpendOnApp (adjust thresh_pct if you need)
_df = df.copy()
_df['HourSpendOnApp'] = group_small(_df['HourSpendOnApp'], thresh_pct=0.02)

# --- PLOTTING ---
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=figsize)
axs = axs.flatten()
cmap = plt.get_cmap('tab20')

for i, var in enumerate(cat_vars):
    ax = axs[i]

    # counts and percentages
    counts_series = _df[var].value_counts()
    if counts_series.sum() == 0:
        ax.text(0.5, 0.5, f"No data for {var}", ha='center', va='center', fontsize=14, color='red')
        ax.axis('off')
        continue

    counts = counts_series.values
    labels = counts_series.index.astype(str).tolist()
    total = counts.sum()
    pcts = (counts / total) * 100

    # explode for very small slices (must match length of counts)
    explode = [0.08 if pct < explode_threshold_pct else 0.0 for pct in pcts]

    # colors matching number of slices
    colors = [cmap(j % cmap.N) for j in range(len(counts))]

    # autopct: only show inside % for slices >= min_pct_inside
    def autopct_func(pct):
        return f'{pct:.1f}%' if pct >= min_pct_inside else ''

    # Draw pie WITHOUT inline labels (we will use legend for all labels)
    wedges, texts, autotexts = ax.pie(
        counts,
        labels=None,            # turn off slice labels to avoid overlap
        autopct=autopct_func,
        startangle=90,
        explode=explode,
        colors=colors,
        pctdistance=0.72,
        radius=0.85,
        wedgeprops={'linewidth': 0.6, 'edgecolor': 'white'},
        textprops={'fontsize': 12}
    )

    # Style visible percent texts inside slices
    for at in autotexts:
        if at.get_text():
            at.set_color('white')
            at.set_fontweight('bold')
            at.set_fontsize(11)

    # Build legend entries (always visible, no overlap)
    legend_labels = [f"{lab}  {pct:.1f}% ({cnt})" for lab, pct, cnt in zip(labels, pcts, counts)]

    # Place legend to the right of each subplot
    ax.legend(wedges, legend_labels,
              title=f"{var} (count)",
              loc='center left',
              bbox_to_anchor=(1.02, 0.5),
              fontsize=10,
              title_fontsize=11,
              frameon=False)

    ax.set_title(f"{var} Distribution", fontsize=16, fontweight='bold', pad=18)
    ax.axis('equal')  # keep pie circular

# Make room on right for the legends and avoid clipping
plt.subplots_adjust(left=0.03, right=0.80, top=0.95, bottom=0.03, wspace=0.35, hspace=0.35)
plt.show()
```

```python
import joblib
import pandas as pd
import numpy as np
import gradio as gr
import os
from datetime import datetime
import matplotlib.pyplot as plt

# ---------------------
# RANDOM CUSTOMER ID GENERATOR
# ---------------------
import random
import string

def generate_random_id():
    part = ''.join(random.choices(string.ascii_uppercase + string.digits, k=8))
    return f"CUST-{part}"

def generate_unique_random_id():
    df = read_master()
    existing_ids = set(df["CustomerID"].astype(str))

    while True:
        new_id = generate_random_id()
        if new_id not in existing_ids:
            return new_id


# ---------------------
# CONFIG / PATHS (AUTO-DETECT)
# ---------------------
# Candidate locations where your training file might be
possible_paths = [
    "/content/E_comm.xlsx",
    "/content/sample_data/E_comm.xlsx",
    "/mnt/data/E_comm.xlsx",
    "/mnt/data/E Commerce Dataset.xlsx",
    "E_comm.xlsx",
    "./E_comm.xlsx"
]

DATASET_PATH = None
for p in possible_paths:
    if os.path.exists(p):
        DATASET_PATH = p
        print(" Training dataset found at:", DATASET_PATH)
        break

if DATASET_PATH is None:
    # If not found, leave None and UI will show helpful message
    print(" E_comm.xlsx NOT FOUND in common locations. Please upload it to the notebook.")

# This is the local path
TRAINING_FILE_LOCAL_PATH = "/content/sample_data/E_comm.xlsx"

# ---------------------
# PERMANENT STORAGE IN GOOGLE DRIVE
# ---------------------
MASTER_DIR = "/content/drive/MyDrive/ChurnCRM"
os.makedirs(MASTER_DIR, exist_ok=True)

MASTER_PATH = f"{MASTER_DIR}/all_customers.csv"

# ---------------------
# AUTO-INIT MASTER CSV (Permanent Fix)
# ---------------------
MASTER_COLUMNS = [
    "Timestamp","CustomerID","Tenure","PreferredLoginDevice","CityTier",
    "WarehouseToHome","PreferredPaymentMode","Gender","HourSpendOnApp",
    "NumberOfDeviceRegistered","PreferedOrderCat","SatisfactionScore",
    "MaritalStatus","NumberOfAddress","Complain",
    "OrderAmountHikeFromlastYear","CouponUsed","OrderCount",
    "DaySinceLastOrder","CashbackAmount","PredictedChurn",
    "ChurnProbability"
]

if not os.path.exists(MASTER_PATH):
    pd.DataFrame(columns=MASTER_COLUMNS).to_csv(MASTER_PATH, index=False)
    print(" Master CSV created automatically:", MASTER_PATH)
else:
    df_master = pd.read_csv(MASTER_PATH, dtype=object)
    changed = False
    for col in MASTER_COLUMNS:
        if col not in df_master.columns:
            df_master[col] = np.nan
            changed = True
    if changed:
        df_master.to_csv(MASTER_PATH, index=False)
        print(" Master CSV verified & fixed (missing columns added):", MASTER_PATH)
    else:
        print(" Master CSV exists and looks fine:", MASTER_PATH)


# ---------------------
# LOAD MODELS (make sure these files exist in working dir)
# ---------------------
class ChurnPredictor:
    def __init__(self):
        # Define model file names
        self.model_files = {
            "decision_tree": "decision_tree_model.pkl",
            "random_forest": "random_forest_model.pkl",
            "xgboost": "xgboost_model.pkl"
        }
        self.dt = None
        self.rf = None
        self.xgb = None
        self.models_loaded = False

        # Check if all model files exist
        all_files_exist = True
        for model_name, file_path in self.model_files.items():
            if not os.path.exists(file_path):
                print(f" Model file not found: {file_path}. Please ensure all model training and saving cells are run first.")
                all_files_exist = False
                break

        if all_files_exist:
            try:
                self.dt = joblib.load(self.model_files["decision_tree"])
                self.rf = joblib.load(self.model_files["random_forest"])
                self.xgb = joblib.load(self.model_files["xgboost"])
                self.models_loaded = True
                print(" All prediction models loaded successfully.")
            except Exception as e:
                print(f" Error loading models: {e}. Please check model files and rerun model saving cells.")
        else:
            print(" Prediction models not loaded due to missing files. Prediction functionality will be disabled.")


    def preprocess(self, df):
        mapping_device = {'Mobile Phone': 0, 'Computer': 1, 'Tablet': 2}
        mapping_payment = {
            'Debit Card': 0, 'UPI': 1, 'Cash on Delivery': 2,
            'E wallet': 3, 'Credit Card': 4, 'Gift Card': 5
        }
        mapping_gender = {'Female': 0, 'Male': 1}
        mapping_order = {
            'Laptop & Accessory': 0, 'Mobile Phone': 1, 'Others': 2,
            'Fashion': 3, 'Grocery': 4, 'Tablet': 5
        }
        mapping_marital = {'Single': 0, 'Married': 1}

        for col, mapping in {
            'PreferredLoginDevice': mapping_device,
            'PreferredPaymentMode': mapping_payment,
            'Gender': mapping_gender,
            'PreferedOrderCat': mapping_order,
            'MaritalStatus': mapping_marital
        }.items():
            if col in df.columns:
                df[col] = df[col].map(mapping)

        df = df.fillna(0)
        return df

    def predict(self, input_data):
        if not self.models_loaded:
            raise RuntimeError("Prediction models are not loaded. Cannot make predictions.")
        df = pd.DataFrame([input_data])
        df = self.preprocess(df)
        preds = [
            self.dt.predict(df)[0],
            self.rf.predict(df)[0],
            self.xgb.predict(df)[0]
        ]
        return int(np.round(np.mean(preds)))

    def predict_proba(self, input_data):
        if not self.models_loaded:
            raise RuntimeError("Prediction models are not loaded. Cannot calculate probabilities.")
        df = pd.DataFrame([input_data])
        df = self.preprocess(df)
        probs = [
            self.dt.predict_proba(df)[0][1],
            self.rf.predict_proba(df)[0][1],
            self.xgb.predict_proba(df)[0][1]
        ]
        return float(np.mean(probs))

predictor = ChurnPredictor()

# ---------------------
# GLOBAL: last predicted record (for current-customer-only graphs)
# ---------------------
LAST_PREDICTION_RECORD = None  # will be set to dict after each predict call

# ---------------------
# UTILS (MASTER CSV)
# ---------------------
def read_master():
    if not os.path.isfile(MASTER_PATH):
        # If master somehow removed, reinitialize
        pd.DataFrame(columns=MASTER_COLUMNS).to_csv(MASTER_PATH, index=False)
    return pd.read_csv(MASTER_PATH, dtype=object)


def get_last_customer_id():
    df = read_master()
    if "CustomerID" in df.columns and len(df) > 0:
        vals = pd.to_numeric(df["CustomerID"], errors="coerce")
        if vals.dropna().size == 0:
            return 0
        return int(vals.max())
    return 0


def save_record_to_master(record: dict):
    df_master = read_master()
    df_new = pd.DataFrame([record])

    for c in df_new.columns:
        if c not in df_master.columns:
            df_master[c] = np.nan

    for required in ["PredictedChurn", "ChurnProbability", "Timestamp", "CustomerID"]:
        if required not in df_master.columns:
            df_master[required] = np.nan

    if "PredictedChurn" in df_new.columns:
        try:
            df_new["PredictedChurn"] = pd.to_numeric(df_new["PredictedChurn"], errors="coerce").astype('Int64')
        except:
            pass

    if "ChurnProbability" in df_new.columns:
        try:
            df_new["ChurnProbability"] = pd.to_numeric(df_new["ChurnProbability"], errors="coerce")
        except:
            pass

    df_master = pd.concat([df_master, df_new], ignore_index=True, sort=False)
    df_master.to_csv(MASTER_PATH, index=False, na_rep='')


# ---------------------
# FORMATTING HELPERS
# ---------------------
def _is_nan_like(x):
    try:
        return pd.isna(x) or str(x).strip() == "" or str(x).lower() == "nan"
    except:
        return False


def risk_level(prob_percent):
    try:
        p = float(prob_percent)
    except:
        p = 0.0
    if p >= 70:
        return "High", "#ff6b6b"
    elif p >= 30:
        return "Medium", "#f7d060"
    else:
        return "Low", "#6bd28e"


def make_colored_card_html(record):
    # Build a compact nice card for predicted records
    prob_raw = record.get("ChurnProbability", "")
    prob = 0.0
    is_predicted = False
    try:
        if not _is_nan_like(prob_raw):
            prob = float(prob_raw)
            is_predicted = True
    except:
        prob = 0.0

    level, color = risk_level(prob)

    pred_val = record.get("PredictedChurn", "")
    tag = "Predicted" if (not _is_nan_like(pred_val)) else "Actual"

    try:
        churn_display = int(float(pred_val)) if not _is_nan_like(pred_val) else ""
    except:
        churn_display = pred_val if not _is_nan_like(pred_val) else ""

    ts = record.get("Timestamp") if (record.get("Timestamp") and not _is_nan_like(record.get("Timestamp"))) else ""

    html = f"""
    <div style="max-width:720px;margin:auto;padding:18px;border-radius:12px;
                box-shadow:0 8px 30px rgba(0,0,0,0.08);font-family:Poppins,Arial;">
      <div style="display:flex;align-items:center;justify-content:space-between;">
        <div>
          <h3 style="margin:0;color:#333;">Customer ID: <b>{record.get("CustomerID")}</b></h3>
          <p style="margin:6px 0 0;color:#555;font-size:14px;">{ts}</p>
        </div>
        <div style="text-align:right">
          <div style="background:{color};padding:8px 12px;border-radius:10px;color:#111;font-weight:700;">
            {level} RISK
          </div>
          <div style="margin-top:8px;font-size:14px;color:#333;">
            <b>{(prob if is_predicted else 0.0):.2f}%</b>
            <div style="height:8px;background:#eee;border-radius:8px;margin-top:6px;overflow:hidden;">
              <div style="width:{(prob if is_predicted else 0.0):.2f}%;height:8px;background:linear-gradient(90deg,#8de6cf,#f9d976,#f27c7c);"></div>
            </div>
          </div>
        </div>
      </div>

      <hr style="border:none;border-top:1px solid #f0f0f0;margin:12px 0;">
      <div style="display:grid;grid-template-columns:repeat(2,1fr);gap:8px;font-size:14px;color:#333;">
        <div><b>Tenure:</b> {record.get('Tenure','')}
</div>
        <div><b>CityTier:</b> {record.get('CityTier','')}
</div>
        <div><b>Device:</b> {record.get('PreferredLoginDevice','')}
</div>
        <div><b>Payment:</b> {record.get('PreferredPaymentMode','')}
</div>
        <div><b>Satisfaction:</b> {record.get('SatisfactionScore','')}
</div>
        <div><b>Complaints:</b> {record.get('Complain','')}
</div>
        <div><b>Orders:</b> {record.get('OrderCount','')}
</div>
        <div><b>HourSpendOnApp:</b> {record.get('HourSpendOnApp','')}
</div>
      </div>
    </div>
    """
    return html


# ---------------------
# TRAINING & PREDICTED HELPERS
# ---------------------
def show_training_customers():
    if DATASET_PATH is None:
        return "<h3>Training dataset not found. Please upload E_comm.xlsx to the notebook.</h3>"
    try:
        df = pd.read_excel(DATASET_PATH)
    except Exception as e:
        return f"<h3>Training dataset could not be read: {e}</h3>"
    return df.to_html(index=False)


def show_predicted_customers():
    df = read_master()

    # Customers are predicted if PredictedChurn has a valid value
    df["PredictedChurn"] = pd.to_numeric(df["PredictedChurn"], errors="coerce")
    predicted = df[df["PredictedChurn"].notna()]

    if len(predicted) == 0:
        return "<h3>No predicted customers yet.</h3>"

    return df_to_html(predicted)


# ---------------------
# PREDICT
# ---------------------
def predict_churn(
    Tenure, PreferredLoginDevice, CityTier, WarehouseToHome,
    PreferredPaymentMode, Gender, HourSpendOnApp, NumberOfDeviceRegistered,
    PreferedOrderCat, SatisfactionScore, MaritalStatus, NumberOfAddress,
    Complain, OrderAmountHikeFromlastYear, CouponUsed, OrderCount,
    DaySinceLastOrder, CashbackAmount
):
    global LAST_PREDICTION_RECORD

    if not predictor.models_loaded:
        return "<h3 style='color:red;'> Prediction models are not loaded. Please re-run the model training and saving cells (above) first.</h3>"

    CustomerID = generate_unique_random_id()


    if OrderAmountHikeFromlastYear is None:
        OrderAmountHikeFromlastYear = 0

    sat_val = int(str(SatisfactionScore).split("")[0].strip()) if isinstance(SatisfactionScore, str) else int(SatisfactionScore)

    input_data = {
        'Tenure': Tenure,
        'PreferredLoginDevice': PreferredLoginDevice,
        'CityTier': CityTier,
        'WarehouseToHome': WarehouseToHome,
        'PreferredPaymentMode': PreferredPaymentMode,
        'Gender': Gender,
        'HourSpendOnApp': HourSpendOnApp,
        'NumberOfDeviceRegistered': NumberOfDeviceRegistered,
        'PreferedOrderCat': PreferedOrderCat,
        'SatisfactionScore': sat_val,
        'MaritalStatus': MaritalStatus,
        'NumberOfAddress': NumberOfAddress,
        'Complain': Complain,
        'OrderAmountHikeFromlastYear': OrderAmountHikeFromlastYear,
        'CouponUsed': CouponUsed,
        'OrderCount': OrderCount,
        'DaySinceLastOrder': DaySinceLastOrder,
        'CashbackAmount': CashbackAmount
    }

    try:
        pred = predictor.predict(input_data)
        prob = predictor.predict_proba(input_data)
        prob_percent = round(prob * 100, 2)
    except RuntimeError as e:
        return f"<h3 style='color:red;'> Prediction Error: {e}</h3>"

    record = {
        "Timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "CustomerID": CustomerID,
        **input_data,
        "PredictedChurn": int(pred),
        "ChurnProbability": prob_percent
    }

    save_record_to_master(record)

    # set current-prediction global for graphs
    LAST_PREDICTION_RECORD = record.copy()

    html_card = make_colored_card_html(record)
    html_table = pd.DataFrame([record]).to_html(index=False)
    return html_card + "<br>" + html_table


# ---------------------
# TABLES & FILTERS
# ---------------------
def df_to_html(df):
    df2 = df.copy()
    # keep UI clean: remove any internal-only columns if present
    if "Churn" in df2.columns:
        # hide actual churn in main tables (still present in training file)
        df2 = df2.drop(columns=["Churn"])
    if "PredictedChurn" in df2.columns:
        df2["PredictedChurn"] = df2["PredictedChurn"].apply(lambda x: "" if _is_nan_like(x) else int(float(x)))
    if "ChurnProbability" in df2.columns:
        df2["ChurnProbability"] = df2["ChurnProbability"].apply(lambda x: "" if _is_nan_like(x) else f"{float(x):.2f}")
    if "Timestamp" in df2.columns:
        df2["Timestamp"] = df2["Timestamp"].apply(lambda x: x if not _is_nan_like(x) else "")
    return df2.to_html(index=False)


def show_all_customers():
    df = read_master()
    return df_to_html(df)


def show_high_risk():
    df = read_master()
    df["ChurnProbability"] = pd.to_numeric(df["ChurnProbability"], errors="coerce")
    return df_to_html(df[df["ChurnProbability"] >= 70])


def show_medium_risk():
    df = read_master()
    df["ChurnProbability"] = pd.to_numeric(df["ChurnProbability"], errors="coerce")
    return df_to_html(df[(df["ChurnProbability"] >= 30) & (df["ChurnProbability"] < 70)])


def show_low_risk():
    df = read_master()
    df["ChurnProbability"] = pd.to_numeric(df["ChurnProbability"], errors="coerce")
    return df_to_html(df[df["ChurnProbability"] < 30])


def show_newly_added():
    df = read_master()

    # Newly added = IDs that are NOT numeric (random customer IDs)
    mask = df["CustomerID"].apply(lambda x: not str(x).isdigit())
    return df_to_html(df[mask])


def show_with_complaints():
    df = read_master()
    return df_to_html(df[pd.to_numeric(df["Complain"], errors="coerce") == 1])


def show_low_satisfaction():
    df = read_master()
    return df_to_html(df[pd.to_numeric(df["SatisfactionScore"], errors="coerce") <= 2])


def show_today():
    today_str = datetime.now().strftime("%Y-%m-%d")
    df = read_master()
    mask = df["Timestamp"].astype(str).str.contains(today_str, na=False)
    return df_to_html(df[mask])


# ---------------------
# SEARCH (FINAL: check Master then Training)
# ---------------------
def search_by_id(search_id):
    search_id = str(search_id).strip()

    # 1 Search in MASTER CSV (predicted customers)
    df_master = read_master()
    df_master["CustomerID"] = df_master["CustomerID"].astype(str)

    sel_master = df_master[df_master["CustomerID"] == search_id]

    if len(sel_master) > 0:
        record = sel_master.iloc[-1].to_dict()
        return """
            <h3 style='color:green;'> Found in Predicted Customers (MASTER CSV)</h3>
        """ + make_colored_card_html(record) + "<br>" + df_to_html(sel_master)

    # 2 Search in TRAINING DATASET
    if DATASET_PATH is not None:
        try:
            df_train = pd.read_excel(DATASET_PATH)
            df_train["CustomerID"] = df_train["CustomerID"].astype(str)
            sel_train = df_train[df_train["CustomerID"] == search_id]
        except:
            sel_train = pd.DataFrame()
    else:
        sel_train = pd.DataFrame()

    if len(sel_train) > 0:
        # Show training table (includes actual churn column if present)
        return """
            <h3 style='color:blue;'> Found in Training Dataset (E_comm.xlsx)</h3>
        """ + sel_train.to_html(index=False)

    # 3 Not found anywhere
    return f"<h3>Customer ID {search_id} not found in database.</h3>"


def export_master():
    return MASTER_PATH


# ---------------------
# BASIC GRAPHS (CURRENT CUSTOMER ONLY)
# ---------------------
def _plot_no_current(ax):
    ax.text(0.5, 0.5, "No current prediction yet.\nPress Predict Churn to generate a chart.",
            horizontalalignment='center', verticalalignment='center', fontsize=12)
    ax.set_xticks([])
    ax.set_yticks([])


def plot_satisfaction_current():
    """
    Plot only the latest predicted customer's SatisfactionScore as a single bar.
    If no current prediction exists, show a message.
    """
    fig, ax = plt.subplots(figsize=(6,4))
    if LAST_PREDICTION_RECORD is None:
        _plot_no_current(ax)
        plt.tight_layout()
        return fig

    try:
        val = float(LAST_PREDICTION_RECORD.get("SatisfactionScore", 0))
    except:
        val = 0.0

    ax.bar([1], [val])
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 5.5)
    ax.set_xticks([1])
    ax.set_xticklabels([f"Customer {LAST_PREDICTION_RECORD.get('CustomerID')}"])
    ax.set_ylabel("Satisfaction Score")
    ax.set_title("Current Customer  Satisfaction Score")
    # annotation
    ax.text(1, val + 0.15, f"{val:.1f}", ha='center', va='bottom', fontsize=12, fontweight='bold')
    plt.tight_layout()
    return fig


def plot_order_count_current():
    """
    Plot only the latest predicted customer's OrderCount as a single bar.
    """
    fig, ax = plt.subplots(figsize=(6,4))
    if LAST_PREDICTION_RECORD is None:
        _plot_no_current(ax)
        plt.tight_layout()
        return fig

    try:
        val = float(LAST_PREDICTION_RECORD.get("OrderCount", 0))
    except:
        val = 0.0

    ax.bar([1], [val])
    ax.set_xlim(0, 2)
    # adapt y limit to value
    ax.set_ylim(0, max(5, val * 1.4))
    ax.set_xticks([1])
    ax.set_xticklabels([f"Customer {LAST_PREDICTION_RECORD.get('CustomerID')}"])
    ax.set_ylabel("Order Count")
    ax.set_title("Current Customer  Total Orders")
    ax.text(1, val + max(0.5, val * 0.05), f"{val:.0f}", ha='center', va='bottom', fontsize=12, fontweight='bold')
    plt.tight_layout()
    return fig


def plot_churn_probability_current():
    """
    Plot only the latest predicted customer's churn probability as a single bar (percentage).
    """
    fig, ax = plt.subplots(figsize=(6,4))
    if LAST_PREDICTION_RECORD is None:
        _plot_no_current(ax)
        plt.tight_layout()
        return fig

    try:
        val = float(LAST_PREDICTION_RECORD.get("ChurnProbability", 0.0))
    except:
        val = 0.0

    ax.bar([1], [val])
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 100)
    ax.set_xticks([1])
    ax.set_xticklabels([f"Customer {LAST_PREDICTION_RECORD.get('CustomerID')}"])
    ax.set_ylabel("Churn Probability (%)")
    ax.set_title("Current Customer  Churn Probability")
    ax.text(1, val + 2, f"{val:.2f}%", ha='center', va='bottom', fontsize=12, fontweight='bold')
    plt.tight_layout()
    return fig


# ---------------------
# UI  WITH DESCRIPTIONS
# ---------------------
with gr.Blocks(theme=gr.themes.Soft(primary_hue="teal", secondary_hue="lime")) as demo:

    gr.HTML("""
    <div style="text-align:center;background:linear-gradient(90deg,#e3fdfd,#ffe6fa);
         padding:18px;border-radius:12px;">
      <h1 style="margin:0;">Customer Churn Prediction  CRM Dashboard</h1>
    </div>
    """)

    with gr.Tabs():

        # ======================================================================
        # PREDICT TAB
        # ======================================================================
        with gr.TabItem("Predict"):
            with gr.Row():
                with gr.Column():
                    Tenure = gr.Number(
                        label="Tenure (months)",
                        info="Customer has been associated with the platform for these many months.",
                        minimum=0
                    )

                    PreferredLoginDevice = gr.Dropdown(
                        ['Mobile Phone','Computer','Tablet'],
                        label="Preferred Login Device",
                        info="The most commonly used device by the customer to log into the platform."
                    )

                    CityTier = gr.Dropdown(
                        [1,2,3],
                        label="City Tier",
                        info="Tier-1: Metro cities, Tier-2: Developed cities, Tier-3: Small towns."
                    )

                    WarehouseToHome = gr.Number(
                        label="Warehouse To Home (km)",
                        info="Distance between customer's delivery location and nearest warehouse.",
                        minimum=0
                    )

                    PreferredPaymentMode = gr.Dropdown(
                        ['Debit Card','UPI','Cash on Delivery','E wallet','Credit Card','Gift Card'],
                        label="Preferred Payment Mode",
                        info="Payment method most frequently used by the customer."
                    )

                    Gender = gr.Dropdown(
                        ['Female','Male'],
                        label="Gender",
                        info="Gender of the customer."
                    )

                    HourSpendOnApp = gr.Number(
                        label="Hour Spend On App",
                        info="Average hours spent daily on the app."
                    )

                    NumberOfDeviceRegistered = gr.Number(
                        label="Number Of Device Registered",
                        info="Total number of devices linked with the customer's account.",
                        value=1
                    )

                    PreferedOrderCat = gr.Dropdown(
                        ['Laptop & Accessory','Mobile Phone','Others','Fashion','Grocery','Tablet'],
                        label="Preferred Order Category",
                        info="Most frequently purchased product category."
                    )

                with gr.Column():

                    SatisfactionScore = gr.Dropdown(
                        label="Satisfaction Score (1 to 5 )",
                        choices=["1  - Worst","2  - Poor","3  - Average","4  - Good","5  - Best"],
                        value="3  - Average",
                        info="Customers self-reported satisfaction rating."
                    )

                    MaritalStatus = gr.Dropdown(
                        ['Single','Married'],
                        label="Marital Status",
                        info="Marital status of the customer."
                    )

                    NumberOfAddress = gr.Number(
                        label="Number Of Address",
                        info="How many delivery addresses the customer has saved.",
                        value=1
                    )

                    Complain = gr.Radio(
                        [0,1],
                        label="Complain",
                        info="Whether the customer has raised complaints in the past."
                    )

                    OrderAmountHikeFromlastYear = gr.Number(
                        label="Order Amount Hike From Last Year (%)",
                        info="Percentage increase in spending compared to last year."
                    )

                    CouponUsed = gr.Number(
                        label="Coupon Used (Total)",
                        info="Total coupons used by the customer.",
                        value=0
                    )

                    OrderCount = gr.Number(
                        label="Order Count (Total)",
                        info="Total number of orders placed.",
                        value=0
                    )

                    DaySinceLastOrder = gr.Number(
                        label="Days Since Last Order",
                        info="Number of days since the customer last placed an order.",
                        value=0
                    )

                    CashbackAmount = gr.Number(
                        label="Cashback Amount ()",
                        info="Total cashback earned by the customer.",
                        value=0
                    )

            submit_btn = gr.Button("Predict Churn", variant="primary")
            predict_output = gr.HTML(label="Result")

            submit_btn.click(
                predict_churn,
                inputs=[Tenure, PreferredLoginDevice, CityTier, WarehouseToHome,
                        PreferredPaymentMode, Gender, HourSpendOnApp, NumberOfDeviceRegistered,
                        PreferedOrderCat, SatisfactionScore, MaritalStatus, NumberOfAddress,
                        Complain, OrderAmountHikeFromlastYear, CouponUsed, OrderCount,
                        DaySinceLastOrder, CashbackAmount],
                outputs=predict_output
            )

        # ======================================================================
        # ALL CUSTOMERS TAB
        # ======================================================================
        with gr.TabItem("All Customers"):
            show_all_btn = gr.Button("Show All Customers", variant="secondary")
            all_output = gr.HTML()
            export_btn = gr.Button("Export CSV")
            export_out = gr.Textbox(label="Export Path")
            show_all_btn.click(show_all_customers, outputs=all_output)
            export_btn.click(export_master, outputs=export_out)

        # ======================================================================
        # FILTERS TAB
        # ======================================================================
        with gr.TabItem("Filters"):
            with gr.Row():
                high_btn = gr.Button("High Risk (>=70%)")
                med_btn = gr.Button("Medium Risk (3069%)")
                low_btn = gr.Button("Low Risk (<30%)")
                new_btn = gr.Button("Newly Added")
            with gr.Row():
                comp_btn = gr.Button("Has Complaints")
                low_sat_btn = gr.Button("Low Satisfaction (<=2)")
                today_btn = gr.Button("Today's Predictions")
            filter_out = gr.HTML()

            high_btn.click(show_high_risk, outputs=filter_out)
            med_btn.click(show_medium_risk, outputs=filter_out)
            low_btn.click(show_low_risk, outputs=filter_out)
            new_btn.click(show_newly_added, outputs=filter_out)
            comp_btn.click(show_with_complaints, outputs=filter_out)
            low_sat_btn.click(show_low_satisfaction, outputs=filter_out)
            today_btn.click(show_today, outputs=filter_out)

        # ======================================================================
        # SEARCH TAB
        # ======================================================================
        with gr.TabItem("Search by ID"):
            id_input = gr.Textbox(label="Enter Customer ID", placeholder="e.g. CUST-8A9D2QW3 or 55631")
            search_btn = gr.Button("Search")
            search_out = gr.HTML()
            search_btn.click(search_by_id, inputs=id_input, outputs=search_out)

        # ======================================================================
        # TRAINING & PREDICTED TAB
        # ======================================================================
        with gr.TabItem("Records"):

            gr.Markdown("###  Training Dataset Customers")
            show_train_btn = gr.Button("Show Training Customers")
            training_output = gr.HTML()
            show_train_btn.click(show_training_customers, outputs=training_output)

            gr.Markdown("###  Predicted Customers (New Rows Added)")
            show_pred_btn = gr.Button("Show Predicted Customers")
            predicted_output = gr.HTML()
            show_pred_btn.click(show_predicted_customers, outputs=predicted_output)

        # ======================================================================
        # GRAPHS TAB (CURRENT-CUSTOMER ONLY)
        # ======================================================================
        with gr.TabItem("Graphs"):
            gr.Markdown("### Analytics Charts  Current Predicted Customer Only")
            with gr.Row():
                sat_btn = gr.Button("Satisfaction Score")
                order_btn = gr.Button("Order Count")
                churn_btn = gr.Button("Churn Probability")
            graph_output = gr.Plot()
            sat_btn.click(plot_satisfaction_current, outputs=graph_output)
            order_btn.click(plot_order_count_current, outputs=graph_output)
            churn_btn.click(plot_churn_probability_current, outputs=graph_output)

    gr.HTML("<div style='font-size:12px;color:#666;margin-top:12px;'>Tip: The CSV is live CRM DB. Predict a customer first, then open Graphs  press the chart you want to see for the current customer.</div>")

if __name__ == '__main__':
     demo.launch(debug=True)
```
